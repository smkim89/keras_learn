```python
from matplotlib import pyplot as plt
import numpy as np
from keras.layers import *
from keras.models import *
from keras.layers import *
from keras.utils import *
from collections import Counter
%matplotlib inline
```

```python

x = np.random.uniform(-np.pi, np.pi, 500)

y = np.random.uniform(-1, 1, 500)
#-1 ~ 1값으로 랜덤하게 분류
X = np.asarray([[x[i], y[i]] for i in range(500)])

print(X.shape)
#500개의 행과 2개의 열
```

```python
sine = np.sin(x)
Y = sine < X[:, 1]
#사인파 또는 정현파는 부드러운주기 진동을 나타내는 수학적 곡선 (x에대한 사인값 y)
plt.scatter(x, sine)
plt.scatter(X[:, 0][Y==0], X[:, 1][Y==0])
plt.scatter(X[:, 0][Y==1], X[:, 1][Y==1])
# 특정 열을 선택하기 위한 표현 첫번째 열을 선택, 두번째 열을 선택
# [500, 2] -> [:, 0], [:, 1]
```

```python
Y = to_categorical(Y)
```

```python
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(2,)))
#2, 는 뒤에자리는 데이터의 갯수인데 생각가능
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(2, activation='softmax'))
#마지막 레이어에는 출력 아웃풋이 클래스 갯수 (2개임)

model.summary()

model.compile(loss='categorical_crossentropy', optimizer='sgd')

model.fit(X,Y,epochs=200)
```

```python
score = model.evaluate(X, Y)
print(score)
```

```python
a = [-2, 0]
b = [2, 0]


pred_x = np.vstack((a,b))
print(pred_x)
```

```python
pred = model.predict(pred_x)
print(pred)
```
