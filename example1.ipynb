```python
from matplotlib import pyplot as plt
import numpy as np
from keras.layers import *
from keras.models import *
from keras.layers import *
from keras.utils import *
from collections import Counter
```

```python

x = np.linspace(1, 10, 1000)

print(x)
```

```python
y = 2*x +1
print(y)
```

```python
model = Sequential()
model.add(Dense(1, activation='linear', input_shape=(1,)))
#Dense Layer에 첫번째 param 은 몇개의 out임 10일경우 다음 레이어의 뉴런의 숫자
#피쳐가 5개면 input_shape 에 5를 넣어준다. 현재는 input이 x 한개임.
#뉴런을 더 추가하려면 model.add로 더 추가해준다.
model.summary()
```

```python
model.compile(loss='mse', optimizer='sgd')
#에러 함수와 옵티마이저도 지정한다.

model.fit(x,y ,epochs=20)
#트레이닝 하기
```

```python
pred_x = []
pred_x = np.append(pred_x, 12)
pred_x = np.append(pred_x, 14)


pred_y = model.predict(pred_x)
#pred_x로 값을 예측한다.
print(pred_y)

plt.plot(x,y)
plt.scatter(pred_x,pred_y)
plt.show()
```
